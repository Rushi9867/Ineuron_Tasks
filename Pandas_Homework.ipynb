```python
import pandas as pd 
data = pd.read_csv("AgentPerformance.csv")
data1 = pd.read_csv("Agent_Login_Report.csv")
data.head()
```

```python
data['New_Date'] = pd.to_datetime(data['Date'])
data['Date1'] = data['New_Date'].dt.day
data['Month'] = data['New_Date'].dt.month
data['Year'] = data['New_Date'].dt.year
data['Week'] = data['New_Date'].dt.week
data.head()
```

```python
# 1 .Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
data.groupby(["Week"])['Average Rating'].mean()
```

```python
#2 .Total working days for each agents
data.groupby(['Agent Name'])['Date1'].count()
```

```python
# 3. Total query that you hvae taken
data.groupby('Agent Name')['Total Chats'].sum()
```

```python
#4. total Feedback that you have received 
data.groupby('Agent Name')['Total Feedback'].sum()
```

```python
#5. a agent name who have average rating between 3.5 to 4
data[(data['Average Rating']> 3.5) & (data['Average Rating']< 4.0)]['Agent Name'].unique()
```

```python
#6 . Agent name who have rating lesss then 3.5
data[data['Average Rating'] < 3.5]['Agent Name'].unique()
```

```python
#7 . agent name who have rating more then 4.5
data[data['Average Rating'] > 4.5]['Agent Name'].unique()
```

```python
#8 . how many feedaback agents have received more than 4.5 average
data.where(data['Average Rating'] > 4.5).groupby('Agent Name')['Total Feedback'].sum()
```

```python
data['DARESPT'] = pd.to_datetime(data['Date']+' '+data['Average Response Time'])
data['DARESLT'] = pd.to_datetime(data['Date']+' '+data['Average Resolution Time'])
data
```

```python
# 12 . percentage of chat on which they have received a feedback
data['PCFDBK'] = (data['Total Feedback']*100)/data['Total Chats']
data.groupby('Agent Name')['PCFDBK'].mean()
```
